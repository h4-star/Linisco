```python
# prompt: Using dataframe df_Ventas: sumar todos los totales que la columna number tengan el numero 0 y tambien sumar la columna total completa y dividir el primero por el segundo y mostrar el total_sum, total_total en formato currency y la division en porcentake

# Calculate the sum of 'total' where 'number' is 0
total_sum_number_zero = df_Ventas[df_Ventas['number'] == 0]['total'].sum()
# Calculate the sum of the entire 'total' column
total_total = df_Ventas['total'].sum()

# Perform calculations before formatting as currency
total_black_calculated = (total_sum_number_zero / 1.21) - total_sum_number_zero
total_total_calculated = (total_total / 1.21) - total_total
division_percentage = (total_sum_number_zero / total_total) * 100

# Format the calculated values as currency and percentage
total_black_currency = f"${total_black_calculated:,.2f}"
total_total_currency = f"${total_total_calculated:,.2f}"
division_formatted = f"{division_percentage:.2f}%"


# Display the results
print(f"total iva black: {total_black_currency}")
print(f"total iva total: {total_total_currency}")
print(f"Division: {division_formatted}")
```

```python
# prompt: Using dataframe df_Ventas: cambiar los values de la columna shopNumber: 66220 = Subway Corrientes, 63953 = Subway Lacroze, 72267 = Subway Ortiz, 10019 = Daniel Ortiz, 30036 = Daniel Lacroze, 30038 = Daniel Corrientes

# Create a mapping dictionary for shopNumber values
shop_mapping = {
    '66220': 'Subway Corrientes',
    '63953': 'Subway Lacroze',
    '72267': 'Subway Ortiz',
    '10019': 'Daniel Ortiz',
    '30036': 'Daniel Lacroze',
    '30038': 'Daniel Corrientes'
}

# Use the map function to replace values in the 'shopNumber' column
df_Ventas['shopNumber'] = df_Ventas['shopNumber'].map(shop_mapping)
```

```python
# prompt: Using dataframe df_Ventas: Contar idSaleOrder agrupando por shopNumber y por hora de orderDate y Graficar con colores distintos de linea de tiempo

import pandas as pd
import matplotlib.pyplot as plt

# Convert 'orderDate' to datetime objects
df_Ventas['orderDate'] = pd.to_datetime(df_Ventas['orderDate'])

# Extract the hour from the 'orderDate' column
df_Ventas['order_hour'] = df_Ventas['orderDate'].dt.hour

# Group by shopNumber and order hour, count idSaleOrder
sales_by_hour = df_Ventas.groupby(['shopNumber', 'order_hour'])['idSaleOrder'].count().reset_index()

# Create the plot
plt.figure(figsize=(16, 8))

for shop in sales_by_hour['shopNumber'].unique():
    shop_data = sales_by_hour[sales_by_hour['shopNumber'] == shop]
    plt.plot(shop_data['order_hour'], shop_data['idSaleOrder'],  label=shop, linewidth=6)

plt.xlabel('Hora')
plt.ylabel('Tickets')
plt.title('Ticket por Unidad de Negocio por Hora')
plt.legend()
plt.xticks(range(24)) # Ensure all hours are displayed on x-axis
plt.grid(True)
plt.show()
```

```python
# prompt: Using dataframe df_Ventas: Grafico de torta por local por suma de total, en el titulo sumar la cantidad total de todos los shopNumber y poner en formato currecy

import pandas as pd
import plotly.express as px

# Group data by shopNumber and sum total
grouped_data = df_Ventas.groupby('shopNumber')['total'].sum().reset_index()

# Calculate total sum of 'total' column
total_sum = df_Ventas['total'].sum()

# Create pie chart
fig = px.pie(grouped_data, values='total', names='shopNumber',
             title=f'Total Ventas por Local <br>Total Ventas SIN IVA: ${total_sum/1.21:,.2f}')

# Show chart
fig
```

```python
# prompt: Using dataframe df_Ventas: hacer un grafico con la suma del total de cada paymentmethod y cada shopNumber

import matplotlib.pyplot as plt
import pandas as pd

# Group data by payment method and shop number, summing the total for each group
grouped_data = df_Ventas.groupby(['paymentmethod', 'shopNumber'])['total'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(18, 9))  # Adjust figure size as needed
plt.bar(grouped_data['paymentmethod'], grouped_data['total'])

# Customize plot labels and title
plt.xlabel('Payment Method - Shop Number')
plt.ylabel('Total Sales')
plt.title('Total Sales by Payment Method and Shop Number')
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent labels from overlapping

# Display the plot
plt.show()

```

```python
# prompt: hagamos un grafico de barras que por ejemplo el mes de junio del 2023 tenga el mismo color de barra que junio 2024 y junio 2025 y que cuente la cantidad de idSaleOrder en la df_Ventas

import pandas as pd
# Asegúrate de que la columna 'orderDate' sea de tipo datetime
df_Ventas['orderDate'] = pd.to_datetime(df_Ventas['orderDate'])

# Extraer el mes y el año
df_Ventas['mes_año'] = df_Ventas['orderDate'].dt.to_period('M')

# Agrupar por mes y año y contar la cantidad de idSaleOrder
ventas_por_mes_año = df_Ventas.groupby('mes_año')['idSaleOrder'].count().reset_index()

# Convertir el PeriodIndex a string para graficar
ventas_por_mes_año['mes_año_str'] = ventas_por_mes_año['mes_año'].astype(str)

# Extraer solo el mes de la columna mes_año
ventas_por_mes_año['mes'] = ventas_por_mes_año['mes_año'].dt.month

# Crear el gráfico de barras usando plotly.express
fig = px.bar(ventas_por_mes_año,
             x='mes_año_str',
             y='idSaleOrder',
             color='mes',  # Usar el mes para asignar color
             labels={'mes_año_str': 'Mes y Año', 'idSaleOrder': 'Cantidad de Ventas'},
             title='Cantidad de Ventas por Mes y Año (con color por Mes)')

# Mostrar el gráfico
fig.show()
```
